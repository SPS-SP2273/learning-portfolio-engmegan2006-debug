### Task 2

```python
def rename_my_file(old_file_name):
    month = old_file_name.split('-')[0]
    day_city_extension = old_file_name.split('-')[1]
    day = day_city_extension.split('_')[0]
    city_extension = day_city_extension.split('_')[1]

    new_filename = f'{month_conversion[month]}-{day}-{city_extension}'
    return new_filename

rename_my_file('jul-10_cairo.txt')
```

```python
for person in ['John','Paul','RINO']:       # file/folder name is not caps-sensitive
    path = os.path.join('people', person) 
    if os.path.exists(path):
        print(f'{path} already exists; skipping creation.')
    else:
        os.mkdir(path)
        print(f'Creating new path, {path}.')
```

```python
for person in ['John', 'Paul', 'Ringo']:
    path_to_destination = os.path.join('people', person)
    shutil.copy('spectrum-01.txt', path_to_destination)
    print(f'Copied file to {path_to_destination}.')     #since is copying of a file that already exists into folders that already exist, no need to use os.mkdir
    
```

```python
for person in ['John', 'Paul', 'Ringo']:
    path_to_text = os.path.join('people', person, 'text')   
    if not os.path.exists(path_to_text):
        os.mkdir(path_to_text)              #creation of folder text

    #to move file: define current path and new path, then use shutil.move
    current_path = os.path.join('people', person, 'spectrum-01.txt')
    new_path = os.path.join('people', person, 'text', 'spectrum-01.txt')
    shutil.move(current_path, new_path, )     # in doing so, text file 'spectrum-01.txt' under PERSON is removed, inserted under TEXT folder
    print(f'Moved text file from {current_path} to {new_path}.')
```

After running shutil.move the first time, new_path will become the new current path. cannot rerun the code a second time, as file is no longer located in current_path, so error: file is not found will surface. To reset (if want to edit code and rerun it to check if it works), can switch shutil.move(current_path, new_path) to shutil.move(new_path, current_path) and run the code, then switch back the sequence when updating code.
