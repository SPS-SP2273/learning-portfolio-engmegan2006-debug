### Exercise 1

```python
for number in range(10):
    square = number**2
    if number % 2 == 0:
        continue
    print(square)
```

### Exercise 2

```python
[number**2 for number in range(10) if number % 2 == 1]
```

### Exercise 3

```python
%%timeit
for number in range(10):
    square = number**2
    if number % 2 == 0:
        continue
    print(square)
```

```python
%%timeit
[number**2 for number in range(10) if number % 2 == 1]
```

### Other Exercises

```python
for power in range(5):      #range 5 means index 0-4, not 1-5
    number = 10**power
    print(power, number)
    if number > 5000: 
        break
```

```python
for power in range(5):
    number = 10**power
    if number > 5000: 
        break
    print(power, number)

# the position of if/break statement matters! if only command the break after the print, power = 4 (number > 5000) will still be printed.
```

```python
for power in range(5):
    if power == 3: 
        continue    # continue = skip    -> don't proceed further in the CURRENT loop where i = 3, move on to i = 4 and so on
    number = 10**power
    print(power, number)
```

```python
[number for number in range(5)]
[number**2 for number in range(5)]
```

```python
for number in range(5):
    print(number)
```
