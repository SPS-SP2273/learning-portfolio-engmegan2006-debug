```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y = x**2
y2 = x**3
plt.title('First Plot')
plt.plot(x,y,y2)
plt.scatter(x,y)
plt.xlabel('time(s)')
plt.ylabel('height(m)')
plt.grid(alpha=0.25)          #alpha gives the transparency of gridlines
```

```python
x = np.linspace(0,6,100)        # replace x = np.arange(6): now for x range 0-6, there are 100 data points, hence smmoother curve
y = x**2
y2 = x**3
sin_x = np.sin(x)

plt.title('First Plot')
plt.plot(x,y,label='Person 1')      # if want legend, must have labels for each plotted curve. each plt.plot() contains only x and one y variable, not x and all y variables
plt.plot(x,y2,label='Person 2', color='red',linestyle='dashed',linewidth=2)
plt.scatter(x,y)

plt.twinx()
plt.plot(x,sin_x, label= '$sin^2x$')  #must include r to prevent \ from being read as an escape sequence
                                        #must include $ $ for ^ to be read as a power, not just another symbol)

plt.legend()                        # only sin^2(x) has a legend now; bc all commands below plt.twinx() only applies to the new y axis                       
                                    #should write a separate plt.legend() for curves related to first y axis
plt.xlabel('time(s)')
plt.ylabel('height(m)')
plt.grid(alpha=0.25)          #alpha gives the transparency of gridlines

plt.savefig('Person 2.jpg')
plt.show()                   

```

```python
x = np.linspace(0,6,100)        # replace x = np.arange(6): now for x range 0-6, there are 100 data points, hence smmoother curve
y = x**2
y2 = x**3

x_error, y_error = 0.1,0.75     # error bar gives the magnitude of error,, here is constant

plt.title('First Plot')
plt.xlabel('time(s)')
plt.ylabel('height(m)')
plt.grid(alpha=0.25)  

plt.plot(x,y,label='Person 1')      # if want legend, must have labels for each plotted curve. each plt.plot() contains only x and one y variable, not x and all y variables
plt.plot(x,y2,label='Person 2', color='red',linestyle='dashed',linewidth=2)
plt.errorbar(x,y2, xerr=x_error, yerr=y_error,label='Person 1 with errors', color='green')

plt.legend()                        # only sin^2(x) has a legend now; bc all commands below plt.twinx() only applies to the new y axis                       
                                    #should write a separate plt.legend() for curves related to first y axis
```

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0, 50, 500)  # 500 points from 0 to 50
y = np.exp(-x / 10) * np.sin(x)

plt.plot(x, y)
plt.grid(alpha=.25)
plt.title(r'$y = e^{-x/10}sin(x)$', fontsize = 20)

plt.xlabel('x', color = 'darkorange', fontsize = 15)
plt.ylabel('y', color = 'darkorange', fontsize = 15, rotation = 0)
plt.ylim([-1,1])
plt.show()

```
